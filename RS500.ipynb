```python
## cleans the data by changing the year and the rank from strings into integers
def clean_albums(albums):
    for album in albums:
        album['number'] = int(album['number'])
        album['year'] = int(album['year'])
    return albums

clean_albums(albums)
```

```python
def find_by_album(name_str):
    album_name = list(filter(lambda w: w['album'] == name_str, albums))
    if len(album_name) == 0:
        return None
    return album_name

def find_by_rank(rank):
    rank_list = list(filter(lambda r: r['number'] == rank, albums))
    if len(rank_list) == 0:
        return None
    return rank_list

def find_by_year(year):
    year_list = list(filter(lambda y: y['year'] == year, albums))
    return year_list

def find_by_years(start_date, end_date):
    years_list = list(filter(lambda ys: start_date <= int(ys['year']) <= end_date, albums))
    return years_list

def find_by_ranks(start_rank, end_rank):
    ranks_list = list(filter(lambda rs: start_rank <= int(rs['number']) <= end_rank, albums))
    return ranks_list

def list_titles(albums):
    titles = list(map(lambda t: t['album'], albums))
    return titles

def list_artists(albums):
    artists = list(map(lambda a: a['artist'], albums))
    return artists

from collections import Counter
def popular_artists(albums):
    c = Counter(list_artists(albums))
    artist_max = max(c.keys(), key = lambda k: c[k])
    return (artist_max, c[artist_max])

popular_artists(albums)
#top_artists = [k for k, v in histogram.items() if v == max(histogram.values())]

def popular_word(albums):
    album_names = list_titles(albums)
    words = []
    for album in album_names:
        words.append(album.split())
    merged_list = []
    for word in words:
        merged_list += word
    unique_words = set(merged_list)
    split_words = dict.fromkeys(unique_words, 0)
    for i in merged_list:
        split_words[i] += 1
    top_words = [k for k, v in split_words.items() if v == max(split_words.values())]
    return max(split_words.values())

def albums_by_decade(albums):
    album_years = list(map(lambda y: y['year'], albums))
    decades = {'1950s': 0, '1960s' : 0, '1970s' : 0, '1980s' : 0, '1990s' : 0, '2000s' : 0, '2010s' : 0}
    for y in album_years:
        if int(y) < 1960:
            decades['1950s'] += 1
        elif 1960 <= int(y) < 1970:
            decades['1960s'] += 1
        elif 1970 <= int(y) < 1980:
            decades['1970s'] += 1
        elif 1980 <= int(y) < 1990:
            decades['1980s'] += 1
        elif 1990 <= int(y) < 2000:
            decades['1990s'] += 1
        elif 2000 <= int(y) < 2010:
            decades['2000s'] += 1
        else:
            decades['2010s'] += 1
    return decades

def hist_genre(albums):
    genres = list(map(lambda g: g['genre'], albums))
    un_genres = set(genres)
    split_g = []
    for g in un_genres:
        split_g.append(g.split(', '))
    merged_g = []
    for g in split_g:
        merged_g += g
    return set(merged_g)
```

```python
import re
def clean_words(albums):
    album_names = list_titles(albums)
    words = []
    for album in album_names:
        words.append(album.split())
    merged_list = []
    for word in words:
        merged_list += word
    upper_merged_list = []
    for word in merged_list:
        upper_merged_list.append(word.upper())
    #upper_merged_list = list(filter(str.isalnum, upper_merged_list))
    alnum_words = []
    for word in upper_merged_list:
        alnum_words.append(re.sub("[^0-9a-zA-Z]+", '', word))
    return alnum_words

def popular_word(albums):
    merged_list = clean_words(albums)
    split_words = dict.fromkeys(set(merged_list), 0)
    for i in merged_list:
        split_words[i] += 1
    top_words = [(k,v) for (k,v)in split_words.items() if v == max(split_words.values())]
    return top_words

popular_word(albums)
```

### Change rank and year into integers instead of strings

```python
def str_to_int(songs_dict):
    for song in songs_dict:
        song['rank'] = int(song['rank'])
        song['year'] = int(song['year'])
    return songs_dict

str_to_int(songs_dict)
```

```python
top_500_songs = []

def top_500_songs_list(songs_dict):
    for song in songs_dict:
        top_500_songs.append(song['song'])
    return top_500_songs

top_500_songs_list(songs_dict)
```



### Find which artists have the most songs

```python
#create artist dictionary
def art_dict(songs_dict):
    artists = []
    for song in songs_dict:
        artists.append(song['artist'])
    #create list of unique artists
    un_artists = set(artists)
    artist_dict = dict.fromkeys(un_artists,0)
    #sort the dictionary by artists
    artist_dict = dict(sorted(artist_dict.items()))
    return artist_dict

#add artists into corresponding bucket to calculate # of songs
def add_to_artist_bucket(songs_dict):
    artist_buckets = art_dict(songs_dict)
    # for loop that iterates thru every song in top 500
    for song in songs_dict:
        # for loop that iterates through every artist in artist bucket
        for artist in artist_buckets:
            # if name of the artist is the same as the artist of the top 500 song, add 1 to artist bucket value
            if artist in song['artist']:
                artist_buckets[artist] +=1
    return artist_buckets

#calculate which artist has the most songs in the top 500 songs
def most_songs_top_500(songs_dict):
    filled_artist_buckets = add_to_artist_bucket(songs_dict)
    top_artists = [(k,v) for (k,v)in filled_artist_buckets.items() if v == max(filled_artist_buckets.values())]
    return top_artists

most_songs_top_500(songs_dict)
```
